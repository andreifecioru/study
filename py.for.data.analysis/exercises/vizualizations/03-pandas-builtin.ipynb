Pandas provides a set of built-ins for data-viz tasks, which are just thin wrappers over seaborn or maptplotlib APIs.

```python
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
# this is a "time-series" DF, where the 1st col is a time-stamp
df1 = pd.read_csv("./data/df1.csv", index_col=0, parse_dates=True)

df1.head()
```

```python
df2 = pd.read_csv("./data/df2.csv")

df2.head()
```

```python
df3 = pd.read_csv("./data/df3.csv")

df3.head()
```

The built-in APIs can be called directly on the data-frame:

```python
df1["A"].hist(bins=30)
```

A more widely used syntax is to go via the `plot` APIs:

```python
df1["A"].plot.hist(bins=30)
```

```python
df2.plot.area(alpha=.5, stacked=True)
```

```python

df2.plot.bar(stacked=True)
```

```python
# if x= param is not passed, it is assumed to be the index col
df1.plot.line(y="B", figsize=(12, 3), lw=1)
```

```python
# this is in effect a 3-D plot, where the 3rd dim is represented by the color
df1.plot.scatter(x="A", y="B", c="C", cmap="coolwarm")
```

```python
# the 3rd dim can be represented by size (bubble plot)
df1.plot.scatter(
    x="A", 
    y="B", 
    s=(df1["C"].abs() * 100).to_list(),
    c="C",
    alpha=.4,
    edgecolor="black",
    linewidth=.5,
    cmap="coolwarm",
    figsize=(8, 6)
)
```

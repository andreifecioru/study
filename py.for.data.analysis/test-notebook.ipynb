```python
class Person:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name
    
    def __repr__(self):
        return f"Person({self.first_name}, {self.last_name})"
        
p = Person("Andrei", "Fecioru")
print(p)

print(vars(p))
```

```python
x = np.linspace(0, 100, 101)
y = x ** 2
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, .8, .95])
axes1.set_xlabel("X values")
axes1.set_ylabel("Y values")
axes1.set_title("Linear function")

axes2 = fig.add_axes([1.1, 0, .8, .95])
axes2.set_xlabel("X values")
axes2.set_ylabel("Y values")
axes2.set_title("Non-linear function")

axes1.plot(x, x)
axes2.plot(x, y)
```

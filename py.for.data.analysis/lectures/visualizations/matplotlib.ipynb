```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 20, 50)
y = x ** 2
```

## Functional API

Simple plot rendering:

```python
plt.plot(x, y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Square function")
```

Multi-plots can be rendered on the same canvas:

```python
# args: no of rows, no of cols, plot no.
plt.subplot(1, 2, 1)
plt.plot(x, x ** 3, "r")

plt.subplot(1, 2, 2)
plt.plot(x, x ** 2, "b")
```

## Object-oriented API

The idea is to create `figure` objects, and call methods on these objects.

```python
fig = plt.figure()

# order: left, bottom, width, height
axes = fig.add_axes((0.1, 0.1, 0.8, 0.8))

axes.plot(x, y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Square function")
```

Multi-plot example: insert axes (graph in graph)

```python
fig = plt.figure()

axes1 = fig.add_axes((0.1, 0.1, 0.8, 0.8))
axes2 = fig.add_axes((0.2, 0.5, 0.4, 0.3))

axes1.plot(x, y, "r")
axes2.plot(y, x, "g")
```

Subplots example:

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, "r")
axes[0].set_title("1st plot")

axes[1].plot(y, x, "g")
axes[1].set_title("2nd plot")

# cleanup layout (deals with overlaps)
# it is recommended to place this at the end of the plot work
plt.tight_layout()
```

You can control the figure size and DPI

```python
fig, axes = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(8, 4), # w/h in inches
    dpi=300
)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

## Adding Legends

```python
fig = plt.figure()

axes = fig.add_axes((0, 0, 1, 1))

# adding the `label=` keyword param is required for the legent to work
axes.plot(x, x ** 2, label="X Squared")
axes.plot(x, x ** 3, label="X Cubed")

# show the legend box ( you can choose where to place it)
plt.legend(loc="best")

plt.tight_layout()
```

## Styling the plots

```python
x = np.linspace(0, 10, 11)
y = x ** 2

fig = plt.figure()

axes = fig.add_axes((0, 0, 1, 1))

axes.plot(
    x, y,
    color="purple", # you can pass-in #hex-codes
    linewidth=3, # you can pass in floating numbers (like 0.5 - to make it really thin)
    alpha=0.5, # controls transparency
    linestyle="--", # there are multiple line styles available
    marker="o", # there are multiple marker styles available

    # styling the markers
    markersize=10,
    markerfacecolor="orange",
    markeredgewidth=2,
    markeredgecolor="black"
)
```

## Axes control and plot ranges

```python
x = np.linspace(0, 20, 200)
y = x ** 2

fig = plt.figure()

axes = fig.add_axes((0, 0, 1, 1))
axes.plot(x, y, color="blue", linestyle="--")

# you can zoom-in on sections of the plot
axes.set_xlim((0, 5))
axes.set_ylim((0,30))
```
